```python
# .loc[] method is approximately same as iloc but here you have to enter the  column name that you need 
new_sample_2=data_2.loc[0:5000,("Make","ID","Fuel Type","Horsepower","Year")]#Make","ID","Fuel Type","Horsepower","Year" are  the name of the column 

new_sample_2=new_sample_2.head(20)# I'm printing 20 data frame from the entire data frame 
new_sample_2
```
